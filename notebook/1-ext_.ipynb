```python
## ETRACCION DE DATOS ##

import os
import zipfile

def unzip_files(src_dir, dest_dir):
    
    for item in os.listdir(src_dir):
        # list all files in source directory
        if item.endswith('.zip'): # check if the file is a zip file
            file_path = os.path.join(src_dir, item)
            # get full path of the zip file
            with zipfile.ZipFile(file_path, 'r') as zip_ref:
                zip_ref.extractall(dest_dir) # extract all files to destination directory
                os.rename(os.path.join(dest_dir, zip_ref.namelist()[0]), os.path.join(dest_dir, item.replace('.zip', '.csv'))) # rename the extracted file

unzip_files('../data/external/', '../data/raw/')
```

```python

import pandas as pd

df = pd.DataFrame()
source_dir = '../data/raw/'

# Concatenamos los sub-datasets
for file_name in os.listdir(source_dir):
    if file_name.endswith('.csv'):
        temp_df = pd.read_csv(os.path.join(source_dir, file_name), low_memory=False)
        df = pd.concat([df,temp_df])
        
# Reemplazando los nombres de los features
df = df[list(var_labels.keys())]


# Reemplazando valores de cada feature
for var, values in var_values.items():
    if values:
        df[var] = df[var].map(values)

# Renombrando las columnas
df.rename(columns=var_labels, inplace=True)


# Guardando el dataset
df.to_csv('../data/pre-processed/pre-dataset.csv', index=False)
```
