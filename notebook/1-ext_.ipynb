```python

### ANALISIS DE DATOS ###

# Autor: @maufranzar.com
# Fuente: INEI - EPEN 2023 - LIMA METROPOLITANA

# Descripcion: 

## DICCIONARIO ##


# Features
var_labels = {
    'MES': 'month',
    'C207': 'sex',
    'C208': 'age',
    'C310': 'rol',
    'C311': 'sector',
    'C317': 'n_employees',
    'C335': 'lf_job',
    'C357_I': 'employment_source',
    'C358': 'n_weeks_lf_job',
    'C366': 'education',
    'C377': 'ethnicity',
    'INGTOTP': 'main_src',
    'INGTOT': 'total_src',
}

# Features_values
var_values = {
    'MES': {1:'jan',2:'feb',3:'mar',4:'apr',5:'may',6:'jun',7:'jul',8:'ago',9:'set',10:'oct',11:'nov',12:'dic'},
    'C207': {1:'m',2:'f'},
    'C208': [],
    'C310': {1:'empleador',2:'independiente',3:'empleado',4:'ayudante_familia',5:'ayudante_familia',6:'hogar',7:'practicante',8:'otro'},
    'C311': {1:'militar',2:'admin_pub',3:'emp_pub',4:'service',5:'emp_priv',6:'otro'},
    'C317': {1:'<=20',2:'21-50',3:'51-100',4:'101-500',5:'>=501'},
    'C335': {1:'True',2:'False'},
    'C357_I': {1:'empleador',2:'agencia',3:'bolsa_pub',4:'bolsa_priv',5:'parientes',6:'anuncios',7:'cv_enviado',8:'publicar',9:'clientes',10:'otros'},
    'C358': [],
    'C366': {1:'otro',2:'otro',3:'otro',4:'otro',5:'otro',6:'secundaria',7:'basica_especial',8:'tecnica_trunca',9:'tecnica',10:'uni_trunca',11:'uni_completa',12:'maestria',13:'otro'},
    'C377': {1:'quechua',2:'aymara',3:'amazonico',4:'otro',5:'afroperuano',6:'blanco',7:'mestizo',8:'otro',9:'otro'}, 
    'INGTOTP': [],
    'INGTOT': [], 
}
```

```python
## EXTRACCION DE DATOS ##

import os
import zipfile
import pandas as pd

def unzip_files(src_dir, dest_dir):
    
    for item in os.listdir(src_dir):
        
        # list all files in source directory
        if item.endswith('.zip'): # check if the file is a zip file
            file_path = os.path.join(src_dir, item)
            
            # get full path of the zip file
            with zipfile.ZipFile(file_path, 'r') as zip_ref:
                
                zip_ref.extractall(dest_dir) # extract all files to destination directory
                os.rename(os.path.join(dest_dir, zip_ref.namelist()[0]), os.path.join(dest_dir, item.replace('.zip', '.csv'))) # rename the extracted file

unzip_files('../data/external/', '../data/raw/')


df = pd.DataFrame()
source_dir = '../data/raw/'

# Concatenamos los sub-datasets
for file_name in os.listdir(source_dir):
    if file_name.endswith('.csv'):
        temp_df = pd.read_csv(os.path.join(source_dir, file_name), low_memory=False)
        df = pd.concat([df,temp_df])
```

```python

### DUPLICATES ###
df = df.drop_duplicates()

# Reemplazando los nombres de los features
df = df[list(var_labels.keys())]


# Reemplazando valores de cada feature
for var, values in var_values.items():
    if values:
        df[var] = df[var].map(values)

# Renombrando las columnas
df.rename(columns=var_labels, inplace=True)


# Guardando el dataset
df.to_csv('../data/pre-processed/predataset.csv', index=False)
df
```
